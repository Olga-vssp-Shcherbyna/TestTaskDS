```python
import json

# Dataset loading
with open('Animals in text train dataset.txt', 'r') as file:
    data = json.load(file)

# Check data structure
print(f"Number of records: {len(data)}")
print(data[0])  # First record output
```

```python
from collections import Counter

# Get all texts
texts = [entry['text'] for entry in data]

# Texts length analysis
text_lengths = [len(text.split()) for text in texts]
print(f"Average text length: {sum(text_lengths) / len(text_lengths):.2f} words")

# Animal frequences
animal_entities = [entity[2] for entry in data for entity in entry['entities']]
animal_counts = Counter(animal_entities)
print(animal_counts.most_common())  # Most common animal output
```
